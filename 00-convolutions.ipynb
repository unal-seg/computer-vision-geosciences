#Visualización Convoluciones en imagenes


## Definición Funciones a usar

```
#plt.clf()
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
folder_imgs =  "imgs/general/"
```

```

 
# ejemplo de kernel, en este caso, hacer nitido (sharpen)
kernel = [(-0.5,-1.0,-0.5),(-1.0,7.0,-1.0),(-0.5,-1.0,-0.5)]
 
# mostrar el kernel en una gráfica de matplotlib, se visualizan valores positivos azul, negativos rojo
def mostrarKernel(kernel, title="Visualización Kernel", size=(3,3)):
    """
    Toma una matriz y devuelve una visualización del kernel en un eje de matplotlib
    entra ->  kernel::ndArray
              title::string
              size::tuple(2)
 
    sale ->   matplotlib ax object
    """
 
    f, ax = plt.subplots(figsize=size)
    im = ax.imshow(kernel, cmap="seismic_r")
    ax.set_title(title)
    f.colorbar(im)
    return ax
 
# Filtrar, toma el kernel (filtro) a ser aplicado en la imagen
 
def filtrar(kernel,name='fallas.jpg', title="Filtro", size=(6,6)):
    """
    Toma una matriz y devuelve la convolución con la imagen
    entra ->  kernel::ndArray
              name::string (termina con la extensión del archivo en el directorio)
              title::string
 
    sale ->   matplotlib ax object
    """
    name = folder_imgs+name # para uso en el sistema de directorios en el colab
    fig, ax = plt.subplots(1, figsize=size)
    image = cv2.imread(name,0)    
    #img = cv2.cvtColor(image, 2)    
    image = cv2.filter2D(image, -1, kernel)    
    plt.imshow(image,cmap="gray")
    plt.title(title)
    return ax
```

### Algunos kernels comunes, expresados de forma matricial


```
sharpen = np.array([(-0.5,-1.0,-0.5),
                    (-1.0,7.0,-1.0),
                    (-0.5,-1.0,-0.5)])


sobelV = np.array([(-0.125, 0.0, 0.125),(-0.25,0.0,0.25), (-0.125, 0.0, 0.125)])
sobelH = np.array([(-0.125, 0.0, 0.125),(-0.25,0.0,0.25), (-0.125, 0.0, 0.125)]).T # transpuesta
sobelD = np.array([(-0.125, -0.25, 0.0),
                   (-0.25, 0.0, 0.25), 
                   (0.0, 0.25, 0.125)])


nada = np.array([(0,0,0),
                 (0,1,0),
                 (0,0,0)]) # este kernel no hace nada, kernel identidad?
```

## Visualización de los kernels

```
mostrarKernel(sharpen, "Sharpen")
mostrarKernel(sobelV, "Sobel Vertical")
mostrarKernel(sobelH, "Sobel Horizontal")
mostrarKernel(sobelD, "Sobel Diagonal")
mostrarKernel(nada, "Ningún Filtro")
```

## Visualización de los filtros en las imagenes

```
filtrar(nada, title="Ningun Filtro") # no hacer nada, devuelve misma imagen en ByN 
filtrar(sharpen, title="Nitido")
filtrar(sobelV, title="Sobel Vertical")
filtrar(sobelH, title="Sobel Horizontal")
filtrar(sobelD, title="Sobel Diagonal")
```

### Con otra imagen
